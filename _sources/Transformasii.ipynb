**Transformasi matriks adalah konsep dalam matematika yang digunakan untuk mengubah koordinat atau representasi suatu objek dalam ruang geometri. Ini sering digunakan dalam berbagai bidang seperti grafika komputer, fisika, dan matematika terapan lainnya.**

**Dalam konteks grafika komputer, transformasi matriks sering digunakan untuk mengubah posisi, rotasi, dan skala objek dalam ruang tiga dimensi. Transformasi matriks umumnya melibatkan matriks transformasi yang diterapkan pada vektor posisi objek untuk menghasilkan vektor posisi yang baru setelah transformasi diterapkan.**

**berikut code bentuk visualisasi refleksi sumbu x dan sumbu y**

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
original = np.array([[1, 2, 3], [1, 3, 1]])

reflect_x = np.array([[1, 0], [0, -1]])

reflect_y = np.array([[-1, 0], [0, 1]])

reflected_x = reflect_x @ original

reflected_y = reflect_y @ original

plt.figure(figsize=(10, 5))

plt.subplot(1, 3, 1)
plt.plot(original[0, :], original[1, :], 'bo-', label='Original')
plt.title('Original')
plt.axis('equal')
plt.grid(True)
plt.axhline(0, color='black',linewidth=0.5)
plt.axvline(0, color='black',linewidth=0.5)

plt.subplot(1, 3, 2)
plt.plot(reflected_x[0, :], reflected_x[1, :], 'ro-', label='Reflected X')
plt.title('Reflected over X-axis')
plt.axis('equal')
plt.grid(True)
plt.axhline(0, color='black',linewidth=0.5)
plt.axvline(0, color='black',linewidth=0.5)

plt.subplot(1, 3, 3)
plt.plot(reflected_y[0, :], reflected_y[1, :], 'go-', label='Reflected Y')
plt.title('Reflected over Y-axis')
plt.axis('equal')
plt.grid(True)
plt.axhline(0, color='black',linewidth=0.5)
plt.axvline(0, color='black',linewidth=0.5)

plt.tight_layout()
plt.show()
```
